# Gathering Earthquake Data

After working with the floods data, we thought to expand to other disasters. I chose 'earthquakes' as my query, and labeled articles containing the word 'earthquake' as relevant or not relelvant to a disaster. This proved tricky though - many articles were indeed about earthquakes, but none pertaining to a DISASTER, just an event (small earthquake of 3.7 reported in the Indian Ocean, for example). Our classes for this topic also proved HUGELY unbalanced, which could cause major problems in our modeling. 

```python
import pandas as pd
import requests
```

```python


url = 'https://newsapi.org/v2/everything?'

param = {
    #'country' : 'us',
    'q': 'earthquake',
    'apiKey' : 'e685d6e1420f4882b86d029ed3c1a11d',
    'from': '2019-01-5',
    'to': '2019-04-15',
    'pageSize': 100,
    'language': 'en',
    'sortBy':'relevancy',
    'sources':'the-washington-post, the-new-york-times, reuters, nbc-news, msnbc, google-news, fox-news, cnn, cbs-news, abc-news, associated-press'
}
every_earthquakes = requests.get(url, params = param)


```

```python
every_earthquakes.json()['articles']
```

```python
articles = every_earthquakes.json()['articles']

for page in range(2, 11):
    print(page)
    param['page'] = page
    
    more_earthquakes = requests.get(url, params = param)
    print(more_earthquakes)
    more_earthquakes = more_earthquakes.json()['articles']
    print(len(more_earthquakes))
    articles.extend(more_earthquakes)
    
len(articles)
    
```

```python
import pandas as pd
arts = pd.DataFrame(articles)
```
